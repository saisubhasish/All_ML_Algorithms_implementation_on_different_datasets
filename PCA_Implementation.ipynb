```python
def pc_calculation(x, no):
    pca = {}
    scaler = StandardScaler()
    x_scaled = scaler.fit_transform(x)
    l, x = np.linalg.eig((x_scaled.T @ x_scaled)/(x_scaled.shapeape[0]-1))
    for i in range(no):
        pc = x_scaled @ x[:,1]
        pca[i]=pc
    pca_df = pd.DataFrame(pca)
    return pca_df
```
